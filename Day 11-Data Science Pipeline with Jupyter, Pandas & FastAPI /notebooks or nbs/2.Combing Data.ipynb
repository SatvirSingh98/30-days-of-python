```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data')
```

```python
for filename in os.listdir(DATA_DIR):
    print(filename)
```

```python
path = os.path.join(DATA_DIR, '2021.csv')
df = pd.read_csv(path)

df.head(10)
```

```python
df['Rank']
```

```python
my_dataframes = []
csv_files = [x for x in os.listdir(DATA_DIR) if x.endswith(".csv")]

for filename in csv_files:
    print(filename)
    year = filename.replace('.csv', '')
    csv_path = os.path.join(DATA_DIR, filename)
    this_df = pd.read_csv(csv_path)
    this_df['filename'] = filename
    this_df['year'] = year
    print(this_df.head(n=1))
    my_dataframes.append(this_df)
```

```python
my_entire_df = pd.concat(my_dataframes)  # to join to dataframes
```

```python
CACHE_DIR = os.path.join(BASE_DIR, 'cache')
os.makedirs(CACHE_DIR, exist_ok=True)

dataset = os.path.join(CACHE_DIR, 'movies-ranking-dataset.csv')

my_entire_df.to_csv(dataset, index=False)
```
